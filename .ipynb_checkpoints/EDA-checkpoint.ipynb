```python
def fix_columns(column): # Normalizar los nombres de las columnas (eliminar espacios, establecer minus)
    return str(column).strip().lower().replace(' ', '_')
    
df.columns = list(map(fix_columns, list(df.columns)))
```

```python
# Corregir formato de la columna 'fecha' y crear nuevas columnas para cada componente de la fecha

df.fecha = pd.to_datetime(df.fecha)
df['year'] = df.fecha.dt.year
df['month'] = df.fecha.dt.month
df['month_day'] = df.fecha.dt.day
df['week_day'] = df.fecha.dt.day_name()

df.head()
```

```python
fig, axes = plt.subplots(figsize=(20, 8))
by_year = df.groupby('year').cantidad.sum()
by_year.plot(kind='line')
plt.title('Cantidad de delitos sexuales por año en Colombia (2010 - 2022)')
plt.ylabel('Cantidad')
plt.xlabel('Año')
plt.show()
```

```python
print('Incremento de casos de delito sexual por cada año respecto al anterior:')
for year in by_year.index[:-1]:
    prev = by_year.loc[year]
    curr = by_year.loc[year + 1]
    increase_percentage = round(((curr - prev) * 100) / prev, 2)
    print(f'{year + 1}: {increase_percentage}%')
```

De la anterior figura, se observa que durante todo el período comprendido entre los años 2010 y 2020, hubo un incremento en los delitos sexuales cometidos en Colombia, siendo entre los años 2015 y 2019 el período con mayor número de casos; los casos en el 2015 aumentaron en un 71% aproximadamente respecto al 2014, lo cual es un aumento alarmante. Sería interesante analizar qué factores en Colombia pudieron aumentar tan significativamente este valor y si mantienen la incidencia hasta el 2019 que es último año donde se aumentaron los casos. Desde el 2020, se observa un decremento en la cantidad de delitos sexuales en el país. Es importante resaltar que los presentes datos tienen registros de casos hasta junio del año 2022.

```python
fig, axes = plt.subplots(figsize=(20, 8))
df.groupby(df.fecha.dt.to_period('M')).cantidad.sum().plot(kind='line')
plt.title('Cantidad de delitos sexuales en Colombia (2010 - 2022)')
plt.ylabel('Cantidad')
plt.xlabel('Año')
plt.show()
```

De la anterior figura, se observa que existen picos a principio de cada año: resulta necesario averiguar si esta información es acertada y verificar qué tipos de delitos sexuales se cometen en estos períodos para identificar patrones y diseñar estrategias de prevención. Asímismo, se observa que a mediados de año existen pequeños picos.

```python
from unidecode import unidecode

def normalizing_strings(string):
    if string == '-':
        string = 'No reportado'
    string = str(string)
    return (unidecode(string)).capitalize()

df.departamento = df.departamento.apply(normalizing_strings)
# df.municipio = df.municipio.fillna('No_reportado')
# df.municipio = df.municipio.apply(normalizing_strings)
```

```python
df.departamento.value_counts().sort_values().plot(kind='barh', figsize=(20, 10))
plt.title('Delitos sexuales por departamento en Colombia (2010 - 2022)')
plt.ylabel('Departamento')
plt.xlabel('Cantidad')
plt.show()
```

```python
"""year_department = df.groupby(['year', 'departamento']).cantidad.sum().groupby(level='year').nlargest(3).reset_index(level=0, drop=True)
year_department.unstack(1).plot(kind='bar', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Departamentos con mayor número de delitos sexuales a través de los años')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()"""
```

```python
departments_most_crimes = df[df.departamento.isin(['Cundinamarca', 'Antioquia', 'Santander', 'Valle'])].groupby(['year', 'departamento']).cantidad.sum()
departments_most_crimes.unstack(1).plot(kind='line', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Departamentos con mayor número de delitos sexuales a través de los años')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.genero.unique()
```

```python
df.genero = df.genero.fillna('No reportado')
df.genero = df.genero.apply(normalizing_strings)
df.genero = df.genero.replace('No reporta', 'No reportado')
```

```python
df[df.genero != 'No reportado'].genero.value_counts().plot(kind='bar', figsize=(20, 8))
plt.title('Cantidad de géneros víctimas de delitos sexuales en Colombia')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
year_genero = df.groupby(['year', 'genero']).cantidad.sum()
year_genero.unstack(1).plot(kind='line', figsize=(20, 8))
plt.title('Cantidad de víctimas según el género a través de los años en Colombia (2010-2022)')
plt.xlabel('Year')
plt.ylabel('Cantidad')
plt.legend(title='Género')
plt.show()
```

```python
df.grupo_etario.unique()
```

```python
df.grupo_etario = df.grupo_etario.fillna('No reportado')
df.grupo_etario = df.grupo_etario.apply(normalizing_strings)
df.grupo_etario = df.grupo_etario.replace('No reporta', 'No reportado')
df.grupo_etario = df.grupo_etario.replace('Adolecentes', 'Adolescentes')
```

```python
df.grupo_etario.value_counts().plot(kind='bar', figsize=(20, 8))
plt.title('Cantidad de víctimas según el grupo etario en Colombia')
plt.ylabel('Cantidad')
plt.xlabel('Grupo etario')
plt.xticks(rotation=0)
plt.show()
```

```python
year_grupo_etario = df.groupby(['year', 'grupo_etario']).cantidad.sum()
year_grupo_etario.unstack(1).plot(kind='line', figsize=(20, 8))
plt.title('Cantidad de víctimas según el grupo etario a través de los años en Colombia (2010-2022)')
plt.xlabel('Year')
plt.ylabel('Cantidad')
plt.legend(title='Grupo etario')
plt.show()
```
