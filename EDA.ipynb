```python
def fix_columns(column): # Normalizing column names (delete spaces, lower letters)
    return str(column).strip().lower().replace(' ', '_')
    
df.columns = list(map(fix_columns, list(df.columns)))
```

```python
df.rename(columns={'armas_medios':'armas',
                    'fecha_hecho':'fecha'}, inplace=True) # To replace some column names for better manipulation
```

```python
# Correct date format for 'fecha' column and create columns for year - month - month_day - week_day

df.fecha = pd.to_datetime(df.fecha)
df['year'] = df.fecha.dt.year
df['month'] = df.fecha.dt.month
df['month_day'] = df.fecha.dt.day
df['week_day'] = df.fecha.dt.day_name()

df.head()
```

```python
from unidecode import unidecode

def normalizing_strings(string):
    if string == '-':
        string = 'No reportado'
    string = str(string)
    return (unidecode(string)).capitalize()

df.departamento = df.departamento.apply(normalizing_strings)
# df.municipio = df.municipio.fillna('No_reportado')
# df.municipio = df.municipio.apply(normalizing_strings)
```

```python
df.departamento.value_counts().sort_values().plot(kind='barh', figsize=(20, 10))
plt.title('Sex crimes per department in Colombia')
plt.ylabel('Department')
plt.xlabel('Count')
plt.show()
```

```python
year_department = df.groupby(['year', 'departamento']).cantidad.sum().groupby(level='year').nlargest(3).reset_index(level=0, drop=True)
year_department.unstack(1).plot(kind='bar', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Departamentos con mayor número de delitos sexuales a través de los años')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
departments_most_crimes = df[df.departamento.isin(['Cundinamarca', 'Antioquia', 'Santander', 'Valle'])].groupby(['year', 'departamento']).cantidad.sum()
departments_most_crimes.unstack(1).plot(kind='line', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Departamentos con mayor número de delitos sexuales a través de los años')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.show()
```

```python
df.armas = df.armas.apply(normalizing_strings)
```

```python
df.armas.unique()
```

```python
df.armas.replace(to_replace={'Armas blancas':'Arma blanca / cortopunzante',
                             'Cortantes':'Arma blanca / cortopunzante',
                             'Cortopunzantes':'Arma blanca / cortopunzante'}, inplace=True)
```

```python
df.armas.value_counts().sort_values().plot(kind='barh', figsize=(20, 8))
plt.title('Tipos de armas utilizados en delitos sexuales en Colombia')
plt.ylabel('Tipo de arma')
plt.xlabel('Cantidad')
plt.show()
```

```python
year_armas = df.groupby(['year', 'armas']).cantidad.sum().groupby(level='year').nlargest(3).reset_index(level=0, drop=True)
year_armas.unstack(1).plot(kind='bar', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Tipo de armas utilizadas en delitos sexuales a través de los años en Colombia')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
year_armas = df[df.armas.isin(['Arma blanca / cortopunzante', 'Contundentes', 'No reportado', 'Sin empleo de armas'])].groupby(['year', 'armas']).cantidad.sum()
year_armas.unstack(1).plot(kind='line', figsize=(20, 8))
plt.legend(title='Departamento')
plt.title('Tipo de armas utilizadas en delitos sexuales a través de los años en Colombia')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
df.genero.unique()
```

```python
df.genero = df.genero.fillna('No reportado')
df.genero = df.genero.apply(normalizing_strings)
df.genero = df.genero.replace('No reporta', 'No reportado')
```

```python
df[df.genero != 'No reportado'].genero.value_counts().plot(kind='bar', figsize=(20, 8))
plt.title('Cantidad de géneros víctimas de delitos sexuales en Colombia')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.show()
```

```python
year_genero = df.groupby(['year', 'genero']).cantidad.sum()
year_genero.unstack(1).plot(kind='line', figsize=(20, 8))
plt.title('Cantidad de víctimas según el género a través de los años en Colombia (2010-2022)')
plt.xlabel('Year')
plt.ylabel('Cantidad')
plt.legend(title='Género')
plt.show()
```

```python
df.grupo_etario.unique()
```

```python
df.grupo_etario = df.grupo_etario.fillna('No reportado')
df.grupo_etario = df.grupo_etario.apply(normalizing_strings)
df.grupo_etario = df.grupo_etario.replace('No reporta', 'No reportado')
df.grupo_etario = df.grupo_etario.replace('Adolecentes', 'Adolescentes')
```

```python
df.grupo_etario.value_counts().plot(kind='bar', figsize=(20, 8))
plt.title('Cantidad de víctimas según el grupo etario en Colombia')
plt.ylabel('Cantidad')
plt.xlabel('Grupo etario')
plt.xticks(rotation=0)
plt.show()
```

```python
year_grupo_etario = df.groupby(['year', 'grupo_etario']).cantidad.sum()
year_grupo_etario.unstack(1).plot(kind='line', figsize=(20, 8))
plt.title('Cantidad de víctimas según el grupo etario a través de los años en Colombia (2010-2022)')
plt.xlabel('Year')
plt.ylabel('Cantidad')
plt.legend(title='Grupo etario')
plt.show()
```
